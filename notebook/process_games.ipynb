```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/Olympic_Games_Summary.csv")
df.head()
```

```python
df['start_date'] = pd.to_datetime(df['start_date']).dt.strftime('%Y-%m-%d')
df.head()
```

```python
df['end_date'] = pd.to_datetime(df['end_date']).dt.strftime('%Y-%m-%d')
df.head()
```

```python
df['competition_start_date'] = pd.to_datetime(df['competition_start_date']).dt.strftime('%Y-%m-%d')
df.head()
```

```python
df['competition_end_date'] = pd.to_datetime(df['competition_end_date']).dt.strftime('%Y-%m-%d')
df.head()
```

```python
df.to_csv('../data/processed/Olympic_Games_Summary.csv', index=False)
```
