```python

from country import CountryOperation, GameOperation


nocTable = pd.DataFrame(CountryOperation.search().json())

medalTable = MedalTableOperation.search()

olympicEditionTable = pd.DataFrame(GameOperation.search().json())
olympicEditionTable = olympicEditionTable[['edition_id', 'edition']]

medalTable
```

```python
medalTable, nocTable, olympicEditionTable
```

```python
medalTable = pd.merge(left=medalTable, right=nocTable,
                        left_on='country_noc', right_on='noc', how='left')
medalTable = pd.merge(left=medalTable, right=olympicEditionTable,
                      left_on='edition_id', right_on='edition_id', how='left')
medalTable.drop(columns=['country_noc'], inplace=True)

medalTable
```

```python
nocTable
```

```python
nocTable = pd.merge(left=nocTable, right=medalTable[['noc']],
                    left_on='noc', right_on='noc', how='right')
nocTable.drop_duplicates(subset='noc', inplace=True)
nocTable
```

```python
medalTable['year'] = medalTable['edition'].str[:4]
medalTable['year'] = pd.Categorical(medalTable['year'], ordered=True)
medalTable
```

```python
medalTable[medalTable['noc'] == "CAN"].sort_values(by='year')
```

```python
medalTable.groupby(['year','noc'])['total'].sum().reset_index()
```

```python
medalTable[~(medalTable['edition'].str.contains('winter',case=False)|medalTable['edition'].str.contains('summer',case=False))]
```

```python
import plotly.express as px


medalTableByYear = medalTable.groupby(['year', 'noc'])[['gold','silver','bronze','total']].sum().reset_index()

medalTableByYear[medalTableByYear['noc'] == 'CAN']
```

```python
topFiveHighestTotal = medalTableByYear.groupby(['noc'])['total'].sum(
).reset_index().sort_values(by='total', ascending=False).head(5)

# Create the line plot
fig = px.line(medalTableByYear[medalTableByYear['noc'].isin(topFiveHighestTotal['noc'].tolist())], x='year', y='total', color='noc',
              title='Country Consistency Over Time (Total Medals)',
              labels={'total': 'Total Medals', 'noc': 'Country'})
fig
```
