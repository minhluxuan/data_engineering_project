```python
import pandas as pd

event_res = pd.read_csv('../data/raw/Olympic_Athlete_Event_Details.csv')
event_res.head()
```

```python
null_percentage = event_res['medal'].isnull().mean() * 100
print(f"Tỷ lệ giá trị NULL trong cột 'medal': {null_percentage:.2f}%")
```

```python
# Xóa cột hoàn toàn trùng nhau
event_res = event_res.drop_duplicates()
# # Tỷ lệ null quá cao, tạo bảng chứa medal tham chiếu đến bảng cũ thông qua khóa ngoại là result_id và athlete_id
# result_medal = athlete_res[['edition_id', 'country_noc', 'result_id', 'athlete_id', 'medal']].dropna(subset=['medal'])
# # Xóa trường medal trong bảng ban đầu
# athlete_res = athlete_res.drop(columns=['medal'])

event_res.head()
```

```python
# Xóa trường edition do trùng với edition_id
event_res = event_res.drop(columns=['edition'])
event_res.head()
```

```python
# Lưu DataFrame vào hai file CSV riêng biệt
event_res.to_csv('../data/processed/Olympic_Athlete_Event_Details_Processed.csv', index=False)
# result_medal.to_csv('../data/processed/Olympic_Athlete_Event_Medal_Details.csv', index=False)
```
