```python
year_by_string = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
year_by_number = np.arange(1, 13)

month_mapping = dict(zip(year_by_string, year_by_number))

def process_row(row):
    if pd.isna(row):
        return row
    else:
        row_splitted = row.strip().split()
        day = row_splitted[0]
        month = row_splitted[-1]
        mapped_month = month_mapping.get(month)
        if mapped_month is None:
            return row
        return f"{mapped_month}-{day}"

# Áp dụng hàm xử lý cho từng hàng trong cột
df['start_date'] = df['start_date'].apply(process_row)

df['start_date'] = df.apply(lambda x: f"{x['year']}-{x['start_date']}" if pd.notna(x['start_date']) else None, axis=1)

df['start_date']
```

```python
def process_competition_date(row):
    if pd.isna(row['competition_date']):
        return None, None  # Nếu NaN thì trả về None cho cả start và end
    else:
        parts = row['competition_date'].strip().split('–')  # Tách chuỗi theo ký tự '–'
        
        if len(parts) != 2:  # Kiểm tra nếu không có đúng 2 phần
            return None, None
        
        # Tách phần cho ngày bắt đầu và ngày kết thúc
        start_part = parts[0].strip()
        end_part = parts[1].strip()
        
        # Tách ngày và tháng cho ngày bắt đầu
        if ' ' in start_part:
            day_start, month_start = start_part.rsplit(' ', 1)
        else:
            day_start = start_part.strip()
            month_start = end_part.split()[-1]  # Lấy tháng từ phần kết thúc nếu không có tháng ở đầu
        
        # Tách ngày và tháng cho ngày kết thúc
        if ' ' in end_part:
            day_end, month_end = end_part.rsplit(' ', 1)
        else:
            day_end = end_part.strip()
            month_end = start_part.split()[-1]  # Lấy tháng từ phần bắt đầu nếu không có tháng ở cuối
        
        # Ánh xạ tháng
        mapped_month_start = month_mapping.get(month_start)
        mapped_month_end = month_mapping.get(month_end)
        
        if mapped_month_start is None or mapped_month_end is None:
            return None, None  # Trả về None nếu không ánh xạ được tháng
        
        # Trả về dạng yyyy-mm-dd cho ngày bắt đầu và kết thúc
        competition_date_start = f"{row['year']}-{mapped_month_start}-{day_start.strip()}"
        competition_date_end = f"{row['year']}-{mapped_month_end}-{day_end.strip()}"
        return competition_date_start, competition_date_end

# Áp dụng hàm xử lý để tạo hai cột mới
df['competition_date_start'], df['competition_date_end'] = zip(*df.apply(process_competition_date, axis=1))

# Hiển thị kết quả
print(df[['competition_date', 'competition_date_start', 'competition_date_end']])
```
